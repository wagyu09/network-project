```python
import pandas as pd
import yfinance as yf

WIKI = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
tables = pd.read_html(WIKI, storage_options={"User-Agent": "Mozilla/5.0"})
sp500 = tables[0]
tickers = sp500["Symbol"].astype(str).str.replace(".", "-", regex=False).unique().tolist()


data = yf.download(
    tickers=tickers,            
    period="3mo",
    interval="1d",
    auto_adjust=False,
    group_by="ticker",
    progress=False,
)

print(data)
 
```

```python
out = (
    data
      .stack(0)                          # 컬럼 첫 레벨(티커)을 행으로 이동
      .rename_axis(['date', 'ticker'])   # 인덱스 이름 정리
      .reset_index()[['ticker', 'date', 'Close', 'Volume']] # 필요한 두 컬럼만
      .sort_values(['ticker', 'date'])   # 정렬
      .reset_index(drop=True)
)

```
